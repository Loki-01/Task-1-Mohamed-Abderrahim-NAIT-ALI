```python
import pandas as pd
```

```python
df = pd.read_excel('Dataset for Data Analytics.xlsx')
```

```python
null_before = df['CouponCode'].isnull().sum()
dup_before = df.duplicated().sum()
dup_id_before = df['OrderID'].duplicated().sum()
```

```python
print("Before cleaning:")
```

```python
print("Rows:", len(df))
```

```python
print("Missing values:", df.isnull().sum())
```

```python
print("Duplicate rows:", df.duplicated().sum())
```

```python
print("Duplicate OrderIDs:", df['OrderID'].duplicated().sum())
```

```python
df['CouponCode'] = df['CouponCode'].fillna('NONE')
```

```python
df = df.drop_duplicates()
```

```python
df = df.drop_duplicates(subset=['OrderID'], keep='first')
```

```python
df['Date'] = pd.to_datetime(df['Date']).dt.strftime('%Y-%m-%d')
```

```python
text_cols = ['Product', 'PaymentMethod', 'OrderStatus',
              'ShippingAddress', 'ReferralSource', 'CouponCode']
```

```python
for col in text_cols:
    df[col] = df[col].astype(str).str.strip()
```

```python
df['UnitPrice'] = df['UnitPrice'].round(2)
df['TotalPrice'] = df['TotalPrice'].round(2)
```

```python
dup_id_after = df['OrderID'].duplicated().sum()
bad_dates_after = (df['Date'].str.len() != 10).sum()
```

```python
print("After cleaning:")
```

```python
print("Rows:", len(df))
```

```python
print("Missing values:", df.isnull().sum())
```

```python
print("Duplicate OrderIDs:", dup_id_after)
```

```python
change_log = pd.DataFrame({
    'Change ID': ['CR001', 'CR002', 'CR003', 'CR004', 'CR005', 'CR006'],
    'Description': [
        "Imputed missing 'CouponCode' values with 'NONE' (no coupon applied)",
        "Checked for exact duplicate rows across all columns",
        "Checked OrderID for duplicate unique identifiers",
        "Standardized 'Date' column to ISO 8601 format (YYYY-MM-DD)",
        "Trimmed whitespace in text fields (Product, PaymentMethod, OrderStatus, "
        "ShippingAddress, ReferralSource, CouponCode)",
        "Rounded UnitPrice and TotalPrice to 2 decimal places"
    ],
    'Impact': [
        f"Resolved {null_before} missing records",
        f"{dup_before} duplicate rows found and removed",
        f"{dup_id_before} duplicates found; {dup_id_after} remaining after cleaning",
        f"{bad_dates_after} incorrectly formatted dates remaining",
        "All text fields normalized",
        "Numeric precision standardized to 2 decimals"
    ],
    'Status': ['Resolved'] * 6
})
```

```python
with pd.ExcelWriter('Dataset_for_Data_Analytics_CLEANED.xlsx') as writer:
    df.to_excel(writer, sheet_name='Cleaned_Data', index=False)
    change_log.to_excel(writer, sheet_name='Change_Log', index=False)

print("\nSaved: Dataset_for_Data_Analytics_CLEANED.xlsx (Cleaned_Data + Change_Log sheets)")
```
